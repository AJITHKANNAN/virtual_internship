```python
import pandas as pd

# Read the three separate Excel files into Pandas DataFrames
content_df = pd.read_csv('/content/Content.csv')
reaction_df = pd.read_csv('/content/Reactions.csv')
reaction_types_df = pd.read_csv('/content/ReactionTypes.csv')
```

# Data Cleaning

# Content_df

```python
content_df.head()
```

```python
content_df.shape
```

```python
# Check for missing values
print(content_df.isnull().sum())
```

```python
# removing unnecessary columns
content_df.drop(['Unnamed: 0', 'User ID', 'URL'], axis =1, inplace= True)
```

```python
content_df.isna().sum()
```

```python
content_df.groupby('Category').sum()
```

```python
# replace some values

content_df['Category'] = content_df['Category'].str.replace('"','')
```

```python
content_df.groupby('Category').sum()
```

```python
# Finding and remove duplicated values
```

```python
content_df.duplicated().sum()
```

```python
content_df.drop_duplicates(subset=['Content ID'])
```

```python
content_df = content_df.drop_duplicates(subset=['Content ID'])
```

```python
# Standardize text data
content_df['Type'] = content_df['Type'].str.lower()
content_df['Category'] = content_df['Category'].str.lower()
```

```python
# Rename column name

content_df.rename(columns= {'Type': 'Content_Type'}, inplace= True)

content_df.columns
```

# Reaction_df

```python
reaction_df.head()
```

```python
reaction_df.shape
```

```python
# rename columns name

reaction_df.rename(columns= {'Type': 'Reaction_Type'}, inplace= True)
```

```python
reaction_df.isna().sum()
```

```python
# Since we are trying to find the top 5 categories by score.
# We removing user id, type which has null values
```

```python
reaction_df.drop(['User ID', 'Unnamed: 0'], axis = 1, inplace = True)
reaction_df.dropna(inplace=  True)
```

```python
reaction_df.isna().sum()
```

```python
# Remove duplicates
reaction_df = reaction_df.drop_duplicates(subset=['Content ID'])
```

```python
# Format datetime
# Assuming Datetime column needs formatting
reaction_df['Datetime'] = pd.to_datetime(reaction_df['Datetime'], format='%Y-%m-%d %H:%M:%S')


reaction_df.head()
```

# Reaction_types_df

```python
reaction_types_df.shape
```

```python
reaction_types_df.head(20)
```

```python
# rename columns name

reaction_types_df.rename(columns= {'Type': 'Reaction_Type'}, inplace= True)
```

```python
reaction_types_df.isna().sum()
```

```python
reaction_types_df.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
reaction_types_df.drop_duplicates(subset= ['Reaction_Type'])
```

# Data cleaning completed

```python
print(content_df.columns)
print('\n')
print(reaction_types_df.columns)
print('\n')
print(reaction_df.columns)
print('\n')
```

```python
# now trying to merge the cleaned datasets

# Merge the DataFrames
merged_df = pd.merge(reaction_df, content_df, on='Content ID', how='left')
merged_df = pd.merge(merged_df, reaction_types_df, on='Reaction_Type', how='left')
merged_df.head()
```

```python
merged_df.columns
```

```python

# Final Dataset
merged_df.to_csv('final_dataset_avi.csv',index = False)
```

# EDA Section

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/final_dataset_avi.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

EDA

```python
# Distribution of scores
plt.figure(figsize=(10, 6))
sns.histplot(df['Score'], bins=20, kde=True, color='skyblue', edgecolor='black')
plt.title('Distribution of Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Top performing categories
category_scores = df.groupby('Category')['Score'].sum().reset_index()
top_5_categories = category_scores.nlargest(5, 'Score')

plt.figure(figsize=(12, 6))
sns.barplot(x='Category', y='Score', data=top_5_categories, palette='pastel')
plt.title('Top 5 Performing Categories by Score')
plt.xlabel('Category')
plt.ylabel('Total Score')
plt.xticks(rotation=45)
# plt.grid(axis='y')
plt.show()
```

```python
# Distribution of scores across categories
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Score', data=df, palette='Set3')
plt.title('Distribution of Scores Across Categories')
plt.xlabel('Category')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Sentiment distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='Sentiment', data=df, palette='Pastel1')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
# Boxplot of scores by sentiment
plt.figure(figsize=(8, 6))
sns.boxplot(x='Sentiment', y='Score', data=df)
plt.title('Scores by Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```

```python

# Bar plot of reaction types
plt.figure(figsize=(10, 6))
sns.countplot(x='Reaction_Type', data=df, palette='pastel')
plt.title('Count of Reaction Types')
plt.xlabel('Reaction_Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
